```python
import pandas as pd
dataframe=pd.read_csv("diabetes.csv")
```

```python
dataframe.head()
```

```python
x=dataframe.drop('Outcome',1)
y=dataframe.Outcome
print(dataframe)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=8)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier()
tree=tree.fit(x_train,y_train)
predictedy=tree.predict(x_test)
from sklearn import metrics
accuracy=metrics.accuracy_score(y_test,predictedy)
print(accuracy)
```

```python
from sklearn.tree import export_graphviz
from io import StringIO
```

```python
dotdata=StringIO()
export_graphviz(tree,out_file=dotdata,filled=True,rounded=True,special_characters=True,class_names=['0','1'])
print(dotdata.getvalue())
```

```python
import pydotplus
graph = pydotplus.graph_from_dot_data(dotdata.getvalue())
graph.write_png("tree_for diabetes.png")
```

```python
from IPython.display import Image
Image(graph.create_png())
```

```python
import pandas as pd
dataframe=pd.read_csv("titanic.csv")
dataframe.head()
```

```python
x=dataframe.drop('Survived',1)
y=dataframe.Survived
```

```python
print(x)
```

```python

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5,random_state=50)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier()
tree=tree.fit(x_train,y_train)
predictedy=tree.predict(x_test)
from sklearn import  metrics
accuracy=metrics.accuracy_score(y_test,predictedy)
print(accuracy)
```

```python
from sklearn.tree import export_graphviz
from io import StringIO
```

```python
dotdata=StringIO()
export_graphviz(tree,out_file=dotdata,filled=True,rounded=True,special_characters=True,class_names=['0','1'])
print(dotdata.getvalue())
```

```python
import pydotplus
graph=pydotplus.graph_from_dot_data(dotdata.getvalue())
graph.write_png("tree for titanic.png")
```

```python
from IPython.display import Image 
Image(graph.create_png())
```
