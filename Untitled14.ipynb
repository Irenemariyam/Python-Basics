```python
df=pd.read_csv("cups of coffee vs hours of sleep.csv")
print(df.groupby("Coffee in ml")["sleep in hours"].mean())
fig=plt.bar(df,x=["1","2","3","4","5","6"],y=df.groupby("Coffee in ml")["sleep in hours"].mean())
fig.show()
```

#this graph shows the cup of coffee and hour of sleep of a person.y axis shows the ml of coffee drank by a person and x axis shows the hours of sleep.Drinking more coffee gives us less amount of sleep.
